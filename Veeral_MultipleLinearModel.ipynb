```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'./DSDPartners_Data.csv', parse_dates=True, squeeze=True)
df.head()
```

```python
#Removing BaseorderID since it has no value, same with CustStoreItemTriadID
df = df.drop('BaseorderID', 1)
df = df.drop('CustStorItemTriadID', 1)

#These are Independent Variables
#ConversionResidual, ActualScans, ForecastedScans, QuantOneOrderAgo, PrevShippedQty, TwoOrderAgo, WeightData, TooBig, SlowMoving
X = df[['ConversionResidual','ActualScans', 'ForecastedScans']]
#This is the Dependent Variable
y = df['PropOrderQty']
```

```python
#Split Test/Train Data
X_train, X_test, Y_train, Y_test = train_test_split( X, y, test_size=0.2, random_state=0)
```

```python
#With sklearn
regr = linear_model.LinearRegression()
regr.fit(X,y)

print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)
```
